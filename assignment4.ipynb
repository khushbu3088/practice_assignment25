```python
4. Write a Python Program to Check Armstrong Number?
```

```python
n=int(input("enter the number:"))
digits=str(n)
power=len(digits)
total=sum(int(i)**power for i in digits)
if n==total:
    print(n,"n is armstrong")
else:
    print(n,"n is not armstrong")
    
```
