```python
Write a Python program to convert kilometers to miles?
```

```python
kilometer=float(input("enter the distance in kilometer"))
miles=kilometer*0.621371
print(f"{kilometer} is equal to {miles:.2f} miles")
```

```python
Write a Python program to convert Celsius to Fahrenheit?
```

```python
celsius=float(input("enter the value of celsius "))
fahrenheit=(celsius*1.8)+32
print(f"{celsius} is equal to {fahrenheit:.2f} fahrenheit")
```

```python
Write a Python program to solve quadratic equation?
```

```python
import cmath
a=int(input("enter the num1"))
b=int(input("enter the num2"))
c=int(input("enter the num3"))

dis=(b**2)-(4*a*c)# discriminant

ans1=(-b-cmath.sqrt(dis))/(2*a)
ans2=(-b+cmath.sqrt(dis))/(2*a)

print(ans1)
print(ans2)
```
